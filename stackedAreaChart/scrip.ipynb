```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set_theme()
```

```python
dfFrame = pd.read_csv('stacked_area_chart_data.csv')
dfFrame
```

```python
plt.figure(figsize=(12,6))
labels = ['Diesel', 'Petrol', 'Gas']
plt.stackplot(
    dfFrame['Year'],
    dfFrame['Diesel'],
    dfFrame['Petrol'],
    dfFrame['Gas'],
)
plt.xticks(dfFrame['Year'],rotation = 45)
plt.legend(labels = labels,loc = 'upper left')
plt.ylabel('Number of cars')
plt.show()
```
