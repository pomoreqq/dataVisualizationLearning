```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = pd.read_csv('line_chart_data.csv')
df['newDate'] = pd.to_datetime(df['Date'])
df
```

```python
labels = ['S&P', 'FTSE']
plt.figure(figsize=(20,8))

plt.plot(df['newDate'],
         df['GSPC500'])
plt.plot(df['newDate'],
         df['FTSE100'])

plt.title('Return of indexes')
plt.ylabel('Returns'),
plt.xlabel('Date')
plt.legend(labels=labels)
plt.show()
```

```python
dataFrameSmallerPART = df[(df['newDate'] >= '2008-07-01') & (df['newDate'] <= '2008-12-31')]
dataFrameSmallerPART
```

```python
labels = ['S&P', 'FTSE']
plt.figure(figsize=(20,8))

plt.plot(dataFrameSmallerPART['newDate'],
         dataFrameSmallerPART['GSPC500'])
plt.plot(dataFrameSmallerPART['newDate'],
         dataFrameSmallerPART['FTSE100'])

plt.title('Return of indexes')
plt.ylabel('Returns'),
plt.xlabel('Date')
plt.legend(labels=labels)
plt.show()
```
