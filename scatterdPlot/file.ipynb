```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = pd.read_csv('scatter_data.csv')
df
```

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(df['Area (ft.)'],
            df['Price'],
            alpha=0.6,
            c=df['Building Type'],
            cmap='mako')
plt.legend(*scatter.legend_elements(),
           loc='upper left',
           title='building type')
plt.title('area x price in estate in california')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data = df,
                x = 'Price',
                y = 'Area (ft.)',
                hue=df['Building Type'],
                palette=['black','darkblue','purple','pink','white'],
                s = 100)

plt.show()
```
