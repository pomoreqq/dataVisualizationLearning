```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
dfDataset = pd.read_csv('pie_chart_data.csv')
dfDataset
```

```python
sns.set_theme()
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(dfDataset['Number of Cars'],labels = dfDataset['Engine Fuel Type'].values,
        autopct= '%.2f%%'
        ,textprops= {'size': 'x-large',
                     'fontweight': 'bold',
                     'rotation': 30})

plt.title('cars by engine fuel type', fontsize = 20)
plt.show()
```
