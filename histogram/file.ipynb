```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme()
```

```python
df = pd.read_csv('histogram_data.csv')
df
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['Price'],
        bins= np.arange(118,538,52.5),
        color='#108A99')
plt.title('Distribution of real estate prices', fontsize=14,fontweight='bold')
plt.xlabel('Price ind thousands of dollars')
plt.ylabel('number of properties')

plt.show()
```
